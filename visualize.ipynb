```python
import os
import matplotlib.pyplot as plt

input_data = os.listdir('data/')
print(input_data)
for input_name in input_data[0:1]:
    os.system("SMMAS.exe < data/" + input_name)
    train_detail = open('detail.txt', 'rb').read()
    open('out/train/' + input_name.split('.')[0] + '_SMMAS.txt','wb').write(train_detail)
    path = open('path.txt', 'rb').read()
    open('out/tour/' + input_name.split('.')[0] + '_SMMAS.txt', 'wb').write(path)
    if(os.path.exists('SMMAS_2OPT.exe')):
        os.system("SMMAS_2OPT.exe < data/" + input_name)
        train_detail = open('detail.txt', 'rb').read()
        open('out/train/' + input_name.split('.')[0] + '_SMMAS_2OPT.txt', 'wb').write(train_detail)
        path = open('Path.txt', 'rb').read()
        open('out/tour/' + input_name.split('.')[0] + '_SMMAS_2OPT.txt', 'wb').write(path)
```

```python
import os
import matplotlib.pyplot as plt

data = os.listdir('out/train/')
print(data)
cnt = 0
for name in data:
    file = open('out/train/' + name, 'r')
    file.readline()
    loop_distances = []
    while 1 + 1 == 2:
        line = file.readline()
        if not line:
            break
        line = line.split(' ')
        if(int(line[0]) >= 3000):
            break
        loop_distances.append(float(line[2]))
    plt.plot(range(len(loop_distances)), loop_distances)
    plt.xlabel('num loops')
    plt.ylabel('best distances')
    cnt = 1 - cnt
    if(cnt == 0):
        plt.legend(['SMMAS', 'SMMAS+2OPT'])
        plt.savefig('visualization/train/' + name.split('.')[0] + '.png')
        plt.show()
```

```python
import os

data = os.listdir('out/tour/')
for name in data:
    if (not name == 'XIT1083_SMMAS_2OPT.txt') and name[-7:-4] == 'OPT':
        print(name)
        os.system('Optimize_3opt.exe < out/tour/' + name)
        path = open('newPath.txt', 'rb').read()
        open('out/tour/' + name.split('.')[0] + '_3OPT.txt', 'wb').write(path)
```

```python
data = os.listdir('out/tour/')
for name in data:
    file = open('out/tour/' + name, 'r')
    n = int(file.readline())
    X = []
    Y = []
    for i in range(n):
        line = file.readline()
        if not line:
            break
        x, y = line.split(' ')
        X.append(int(x))
        Y.append(int(y))
    X.append(X[0])
    Y.append(Y[0])
    plt.plot(X, Y)
    plt.scatter(X, Y, s = 4, c = 'r')
    plt.savefig('visualization/tour/' + name.split('.')[0] + '.png')
    plt.show()
```

```python
data = os.listdir('data')
print(data)
for name in data:
    file = open('data/' + name)
    points = []
    while 1 + 1 == 2:
        line = file.readline()
        if not line:
            break
        line = line.split()
        points.append([int(line[1]), int(line[2])])
    
    file = open('optimal_tour/' + name)
    X = []
    Y = []
    file.readline()
    for i in range(len(points)):
        line = file.readline()
        #print(line)
        index = int(line) - 1
        X.append(points[index][0])
        Y.append(points[index][1])
    X.append(X[0])
    Y.append(Y[0])
    plt.plot(X, Y)
    plt.scatter(X, Y, s = 4, c = 'r')
    plt.savefig('visualization/tour/' + name.split('.')[0] + '_OPT.png')
    plt.show()
```
